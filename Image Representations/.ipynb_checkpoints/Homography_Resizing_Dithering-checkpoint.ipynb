### Image Resizing
There are a few ways to interpolate images after resizing.

Each method will contain pixel values from the original image, and a new set of pixels after transformation.

The new pixel values must be 'guessed' through the relationship between a new pixel value, and its neighboring pixels.

There are three methods to 'guess' this pixel value.
1. **Nearest Neighbors**
    - The new pixel values will hold the value of its nearest neighbor pixel. This is the naive method.
    
2. **Bilinear**
    - The new pixel will be a ratio between the distance between that pixel, and 4 neighbors within its neighborhood, these neighbors are in the outer 4 corners of a 3x3 neighborhood. The pixel in location row 2, col 2, will estimate an intensity level for neighbors at location 1,1; 3,1; 1,3; 3,3. 
    
3. **Bicubic**
    - Same idea as bilinear, except the new transformed pixel intensity look at 16 neighbors.

```python
resize_factor = 4
# resize with nearest neighbors interpolation
img1_new1 = img1.resize((int(img1.width/resize_factor), 
                       int(img1.height/resize_factor)), Image.NEAREST)
# resize with bilinear interpolation
img1_new2 = img1.resize((int(img1.width/resize_factor),
                       int(img1.height/resize_factor)), Image.BILINEAR)
# resize with bicubic interpoloation
img1_new3 = img1.resize((int(img1.width/resize_factor), 
                        int(img1.height/resize_factor)), Image.BICUBIC)

# plot images
plt.figure(figsize=(20,16))
plt.subplot(131)
plt.title('Nearest')
display_image(img1_new1)

plt.subplot(132)
plt.title('Bilinear')
display_image(img1_new2)

plt.subplot(133)
plt.title('Bicubic')
display_image(img1_new3)

plt.show()
```

## Dithering
To Dither means to reduce the number of intensity levels, Dithering with Floyd-Steinberg and 1-bit quantization

```python
# 1-bit quantization
img = Image.open('data/graybars.pgm')
binary_img = np.array(img, dtype='float')
binary_img = (binary_img>=128)*255  # binary image conversion

plt.figure(figsize=(20,16))
plt.subplot(121)
plt.title('Original')
display_image(img)

plt.subplot(122)
plt.title('1-bit quantized')
display_image(binary_img)

plt.show()

# Floyd-Steinberg Dithering
im  = np.array(img,dtype=float)
bim = np.zeros(im.shape)
for y in range(1,im.shape[0]-1):    
  for x in range(1,im.shape[1]-1): 
    if im[y,x] >= 128:
      bim[y,x] = 255
    epsilon = im[y,x]-bim[y,x]
    im[y,x+1]   = im[y,x+1]   + 7*epsilon/16
    im[y+1,x-1] = im[y+1,x-1] + 3*epsilon/16
    im[y+1,x]   = im[y+1,x]   + 5*epsilon/16
    im[y+1,x+1] = im[y+1,x+1] + epsilon/16
    
plt.figure(figsize=(20,16))
plt.subplot(121)
plt.title('Original')
display_image(img)

plt.subplot(122)
plt.title('Two levels')
display_image(bim)
plt.show()
```

### Image Homography

choosing four source points, and four target points, transform image by finding the transform matrix H that transforms all source pixels to target pixels

```python
from matplotlib.patches import Circle

# read image
img = Image.open('data/homography_ex1.png')
# display first four points on image
fig,ax = plt.subplots(1)
ax.set_aspect('equal')
display_image(img)

# Select four source points
circ = Circle((175,187),10,color='b')
ax.add_patch(circ)

circ = Circle((510,187),10,color='r')
ax.add_patch(circ)

circ = Circle((45,575),10,color='g')
ax.add_patch(circ)

circ = Circle((590,575),10,color='y')
ax.add_patch(circ)

plt.show()
```

```python
import numpy as np

# Original source points
xp = [175,510,45,590]
yp = [187,187,575,575]
# Target points
x = [10,650,10,650]
y = [10,10,806,806]
# Solving the Inverse Transform
A  = np.zeros([8,8])
for i in range(4):
  A[2*i,:]   = [x[i],y[i],1,0,0,0,-x[i]*xp[i],-y[i]*xp[i]]
  A[2*i+1,:] = [0,0,0,x[i],y[i],1,-x[i]*yp[i],-y[i]*yp[i]]
b = np.array([xp[0],yp[0],xp[1],yp[1],xp[2],yp[2],xp[3],yp[3]])
h = np.linalg.solve(A, b)  

img1 = img.transform((img.width,img.height),Image.PERSPECTIVE,h,Image.BICUBIC)
display_image(img1)
plt.show()
```

In the above code, we solve for the H matrix by first creating A.

Next, we input the values of Matrix A that satisfy the Ax = b equation. Where A is the coefficients of all 8 h parameters. And x are all 8 h parameters, and b are the source x,y points.
